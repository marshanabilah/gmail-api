```python
import marimo as mo
from services_api import init_gmail_service, get_email_message_details, search_emails, search_email_conversations
import json

client_file = "client_secret.json"
service = init_gmail_service(client_file)
```

```python
query = 'from:info@sevenbank.co.jp'
email_messages = search_emails(service, query, max_results=5)

# save the email messages to a JSON file

for message in email_messages:
    email_detail = get_email_message_details(service, message['id'])
    print(message['id'])
    print(f"Subject: {email_detail['subject']}")
    print(f"Date: {email_detail['date']}")
    print(f"Snippet: {email_detail['snippet']}")
    print(f"Labels: {email_detail['label']}")
    print(f"Body: {email_detail['body']}") 
    print(f"Content Type: {email_detail['content_type']}")
    print("-" * 50)

# Save the email details to a JSON file
email_details = [get_email_message_details(service, message['id']) for message in email_messages]
with open('email_details.json', 'w') as f:
    json.dump(email_details, f, indent=4)
```
